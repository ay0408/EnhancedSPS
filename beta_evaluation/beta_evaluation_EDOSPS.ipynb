```python
import numpy as np
import math
from scipy import integrate
import matplotlib.pyplot as plt
import random
import time
```

```python
def TDT(b,c):
    if b+c == 0:
        return 0
    else:
        return ((b-c)**2)/(b+c)
```

```python
def LS(b,c,N):
    stat = TDT(b,c)
    M = stat; m = stat
    
    if b >= 2:
        v = [TDT(b-2,c), TDT(b-2,c+1), TDT(b-2,c+2)]
        M = max([max(v),M])
        m = min([min(v),m])
    if b >= 1:
        v = [TDT(b-1,c), TDT(b-1,c+1)]
        M = max([max(v),M])
        m = min([min(v),m])
    if b >= 1 and c >= 1:
        v = TDT(b-1,c-1)
        M = max([v,M])
        m = min([v,m])
    if c >= 2:
        v = [TDT(b,c-2), TDT(b+1,c-2), TDT(b+2,c-2)]
        M = max([max(v),M])
        m = min([min(v),m])
    if c >= 1:
        v = [TDT(b,c-1), TDT(b+1,c-1)]
        M = max([max(v),M])
        m = min([min(v),m])
    if b+c <= 2*N-2:
        v = [TDT(b+1,c+1), TDT(b,c+2), TDT(b+2,c)]
        M = max([max(v),M])
        m = min([min(v),m])
    if b+c <= 2*N-1:
        v = [TDT(b,c+1), TDT(b+1,c)]
        M = max([max(v),M])
        m = min([min(v),m])
    if b+c <= 2*N-1 and b >= 1:
        v = TDT(b-1,c+2)
        M = max([v,M])
        m = min([v,m])
    if b+c <= 2*N-1 and c >= 1:
        v = TDT(b+2,c-1)
        M = max([v,M])
        m = min([v,m])
    
    return [M-stat, stat-m]
```

```python
def beta_u_EDOSPS(N): # U_r = {x|exists r'. max_y((u(x,r)-u(y,r)) + (u(y,r')-u(x,r'))) > 12} # Approximated
    GS = 8*(N-1)/N
    mb = 10000
    for b1 in range(2*N+1):
        for c1 in range(b1,2*N-b1+1):
            for b2 in range(b1,2*N+1):
                for c2 in range(b2,2*N-b2+1):
                    ls1 = LS(b1,c1,N); ls10 = ls1[0]; ls11 = ls1[1]
                    ls2 = LS(b2,c2,N); ls20 = ls2[0]; ls21 = ls2[1]
                    if ls10 + ls21 <= 12 and b1+c1+b2+c2 != 0:
                        sd = math.ceil((12-ls10-ls21)*(b1+c1)*(b2+c2)/(32*(b1+c1+b2+c2)))+1
                        if ls10 + ls21 != 0:
                            if math.log(2*GS/(ls10+ls21))/sd < mb:
                                print(b1,c1,b2,c2,sd,"01")
                            mb = min([math.log(2*GS/(ls10+ls21))/sd, mb])
                    if ls11 + ls20 <= 12 and b1+c1+b2+c2 != 0:
                        sd = math.ceil((12-ls11-ls20)*(b1+c1)*(b2+c2)/(32*(b1+c1+b2+c2)))+1
                        if ls11 + ls20 != 0:
                            if math.log(2*GS/(ls11+ls20))/sd < mb:
                                print(b1,c1,b2,c2,sd,"10")
                            mb = min([math.log(2*GS/(ls11+ls20))/sd, mb])
                            
    return mb
```
